```python
import requests
from bs4 import BeautifulSoup as bs

url = 'http://www.ss.net.tw/list1.asp?num=136'
head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/48.0.2564.116 Safari/537.36'}

res = requests.get(url, headers = head)

print res
```

```python
soup = bs(res.text) 

print soup
```

```python
ul = soup.select('ul img')

print ul
```

### 抓圖片

```python
for i in ul:
    print i['src']
```

### 抓頁數

```python
a = soup.select('td tr p a')
for i in a:
    print i
```
