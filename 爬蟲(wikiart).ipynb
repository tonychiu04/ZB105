```python
import requests
from bs4 import BeautifulSoup as bs

import time
import json

def grab_list(style,num):
    print style
    head = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/48.0.2564.116 Safari/537.36'}
    url = 'http://www.wikiart.org/en/paintings-by-style/{}?select=featured&json=2&page={}'.format(style,num)
    res = requests.get(url, headers=head)
    print res
    soup = bs(res.text)
    
    content = soup.select('p')[0].text
    ###
    img_l = json.loads(content)
    for i in img_l['Paintings']:
        print i['title']
        print i['year']
        print i['artistName'].strip()
        print i['image']+'!Large.jpg'
        print '-----------------------------------'
    ###
    #save_list(content,num)
    print 'get the data {}'.format(num)
    time.sleep(5)
    
###start###

style = ['abstract-art','abstract-expressionism','art-informel','art-nouveau-modern',
        'baroque','color-field-painting','conceptual-art','cubism','early-renaissance',
         'expressionism','high-renaissance','impressionism','lyrical-abstraction',
         'magic-realism','mannerism-late-renaissanc','minimalism','na-ve-art-primitivism',
         'neoclassicism','neo-expressionism','northern-renaissance','op-art','pop-art',
         'post-impressionism','realism','rococo','romanticism','surrealism','symbolism',
         'ukiyo-e'
        ]
for s in range(1):
    for page in xrange(1,2):
        grab_list(style[2],page)
 
```
