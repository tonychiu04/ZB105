### 偽裝成瀏覽器

```python
import requests

#以卡提諾漫畫網來舉例  (假設網站禁止機器人)


url = 'http://comic.ck101.com'
head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/48.0.2564.116 Safari/537.36'}
#這個主要是模擬瀏覽器來請求網頁
res = requests.get(url, headers = head)
```

### 使用BeautifulSoup 來抓取網頁特定內容

```python
from bs4 import BeautifulSoup as bs

soup = bs(res.text) 
#.text可以把內容全部取出
#.contents 可以把內容以tag的方式放入list中
#.select 可以選定特定物件 ''=標籤, '#'=id, '.'=class，取出來為list


gsoup = soup.select('.hotComic a')#只選取class=hotcomic中tag=a的標籤

#print gsoup[0]#因為是list，所以要用index來看，不然中文會變亂碼


##############################這是分隔線###################################

#這邊要注意，這個class下的a tag有包含漫畫的圖片和名稱
#所以在處理的時候不要重複抓取了
count = 1
for i in gsoup:
    print count
    print i
    count += 1
##############################這是分隔線###################################
    

```
