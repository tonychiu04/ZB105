### Plotting numpy arrays as images

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img=mpimg.imread('0.jpg')
print img
imgplot = plt.imshow(img)
```

### Applying pseudocolor schemes to image plots

```python
lum_img = img[:,:,0]
plt.imshow(lum_img) 
plt.colorbar()
```

```python
plt.imshow(lum_img, cmap="hot")
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('spectral')
plt.colorbar()
```

```python
#可以看顏色的分布圖
plt.hist(lum_img.ravel(), bins=256, range=(0, 240), fc='k', ec='k')
```

```python
imgplot = plt.imshow(lum_img, clim=(0.0, 150.0))
```

### Array Interpolation schemes

```python
from PIL import Image
img = Image.open('0.jpg')
img.thumbnail((64, 64), Image.ANTIALIAS) # resizes image in-place
imgplot = plt.imshow(img)
```

```python
imgplot = plt.imshow(img, interpolation="nearest")
```

```python
imgplot = plt.imshow(img, interpolation="bicubic")
```

### 圖片的讀取和大小

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

from skimage.color import rgb2gray
from skimage import data
from skimage.filters import gaussian
from skimage.segmentation import active_contour
%matplotlib inline

img = plt.imread('2.jpg')
img = rgb2gray(img)

fig = plt.figure(figsize=(4,4)) 
ax = fig.add_subplot(111) #放格子
plt.gray()
ax.imshow(img)  #用格子來放圖
plt.show()
```

### 圖片的裁剪方式

```python
k = plt.imread('2.jpg')[0:100]
plt.imshow(k)
plt.show()
```

```python
plt.hist(k.ravel(), bins=256, range=(0, 255), fc='k', ec='k')
```
